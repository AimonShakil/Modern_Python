## Scalar Value

```python
import numpy as np

a : np.ndarray = np.array(1000) # this is a numpy Object to store, ndarray is numpy class, & we have to look into its basic properties , this 1000 or any other single number is scalar and scalar has no dimension

display(a) # print
display(a.shape) # prints the shape of object () = 0 - Dimension
display(a.dtype) # print the data typpe of the Object 
display(type(a)) # prints 
display(a.ndim) # print the number of the dimensions
display(a.size) # print the size of the object, size tells about number of items in any array 
display(a.itemsize) # print the item size of the object , how many bits it is stored in

# these are basic properties we check on data 
```

## Vector Type

```python

a : np.ndarray = np.array([1, 2, 3, 4]) # this is a vector type, values stored in a list with one dimension

display(f" {a}") # print
display(f" Object shape {a.shape}") # prints the shape of object () = 0 - Dimension (means axis)
display(f" Object data type {a.dtype}") # print the data typpe of the Object 
display(f" Global object type with global function {type(a)} ") # prints 
display(f" Number of Dimension {a.ndim}") # print the number of the dimensions
display(f" Total items in array {a.size}") # print the size of the object, size tells about number of items in any array 
display(a.itemsize) # print the item size of the object , how many bits it is stored in
```

## Matrix

```python
import numpy as np
data = [[1, 2, 3, 4],  # 2d array, we ahave 3 rows and 4 columns, we have a matrix with 3 rows and 4 column
        [5, 6, 7, 8],
        [9, 10, 11, 12]]


a : np.ndarray = np.array(data) # this is a vector type, values stored in a list with one dimension

display(f" {a}") # print
display(f" Object shape {a.shape}") # prints the shape of object (2) = 2 - Dimension (means axis)
display(f" Object data type {a.dtype}") # print the data typpe of the Object 
display(f" Global object type with global function {type(a)} ") # prints 
display(f" Number of Dimension {a.ndim}") # print the number of the dimensions
display(f" Total items in array {a.size}") # print the size of the object, size tells about number of items in any array 
display(a.itemsize) # print the item size of the object , how many bits it is stored in
```

## Numpy with NDArray typing Support

```python
# we have to make an array

from nptyping import NDArray, Shape, UInt64
from typing import Any

data : NDArray[Shape["10"], Any] = np.arange(1 ,11) # here data is a variable with type NDArray , then defining shape means making a matrix with number of items as 10, unsigned Integer thi is data 
                                    # if we want to generate range.


data # a 1 dimensional array has been generated 
```

```python
# matrix of any dimension means 1D, 2D, 5D manipulation on it as if we say + 5 in evrey value

from nptyping import NDArray, Shape, UInt64
from typing import Any

data : NDArray[Shape["10"], Any] = np.arange(1 ,11)

data + 5 # this can not be done through list with + operator directly between 2 operands
```

# Numpy 1d Functions

```python
state_bank : NDArray[Shape["10"], Any] = np.array([1, 5, 7, 12]) # with np function we can convert any tuple, set, lsit can be convert to an Array, from range fixed numbers generte

ubl_bank : NDArray[Shape["100"], Any] = np.random.randint(1,100, 20)   # here we use , with randint we can tell high and low numbers adn its types
                                                                # set here boundry of min and max number and specified number of items to be generated 

display(state_bank)
display(ubl_bank)
```

### Boolean Slicing 

```python
# Boolean Slicing : Topic for Slicing in Numpy
from nptyping import Bool 
state_bank : NDArray[Shape["10"], Any] = np.array([1, 5, 7, 12])
# [1,       5,    7,    12] #according to rule it is must to have every index same dimension, this array is one dimension with 4 index, rhen boolean slicer should also have same dimension and number of item
# [True,  False, false, true]
# variable [True,  False, false, true], it will return where ever have true values 

select : NDArray[Shape["10"], Bool]= np.array([True,  False, False, True])
state_bank[select]
```

```python
# making Boolean Array Method 2 
state_bank : NDArray[Shape["10"], Any] = np.array([1, 5, 7, 12])
state_bank % 2 == 2
```

```python
state_bank : NDArray[Shape["10"], Any] = np.array([1, 5, 7, 12])
state_bank[state_bank % 2 == 2]
```

```python
display(state_bank)
display(ubl_bank)
#        4 Items     search 20 Items
state_bank[np.in1d(state_bank,ubl_bank)]# first varibale item length returned 
```

```python
np.intersect1d(state_bank,ubl_bank) # for unions and diagrams
```

```python
# performing the comparison, needed for comparison between matrices 
x : NDArray[Shape["5"],Any ] = np.array([1, 3, 4, 5, 7])
y : NDArray[Shape["5"],Any ] = np.array([6, 3, 2, 5, 7, 100])
display(x)
display(y)
np.where(x > y, x, y)
```
